```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def apply_pool(img, pool):
    res = np.zeros([img.shape[0]//pool[0], img.shape[1]//pool[1]])
    for r in range(res.shape[0]):
        for c in range(res.shape[1]):
            cut = img[r*pool[0]:r*pool[0]+pool[0], c*pool[1]:c*pool[1]+pool[1]]
            res[r][c] = np.max(cut)
    
    return res
```

```python
img = cv2.imread("strange.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
small = cv2.resize(gray, (gray.shape[1]//2, gray.shape[0]//2))
```

```python
out = apply_pool(small, [2, 2])
```
