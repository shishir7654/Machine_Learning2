```python
from keras.layers import Input,Dense
from keras.models import Sequential,Model
from keras.optimizers import Adam
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("mnist_train_small.csv").values
```

```python
data.shape
```

```python
X = data[:, 2:] / 255
```

```python
def get_adam():
    return Adam(lr=.001)
```

```python
def get_generator():
    adam = get_adam()
    model = Sequential()
    
    model.add(Dense(128 , input_dim =50,activation = 'sigmoid'))
    model.add(Dense(256,activation ='relu'))
    model.add(Dense(512,activation ='relu'))
    model.add(Dense(1024,activation ='relu'))
    model.add(Dense(784,activation ='relu'))
    model.compile(loss = "binary_crossentropy",optimizer =get_adam() , metrics=["accuracy"])
    return model
```

```python
get_generator().summary()
```

```python

def get_discriminator():
    adam = get_adam()
    model = Sequential()
    
    model.add(Dense(128 , input_dim = 784,activation = 'sigmoid'))
    model.add(Dense(256,activation ='relu'))
    model.add(Dense(512,activation ='relu'))
    model.add(Dense(1024,activation ='relu'))
    model.add(Dense(784,activation ='relu'))
    model.add(Dense(1 , activation = 'sigmoid'))
    model.compile(loss = "binary_crossentropy",optimizer =get_adam() , metrics=["accuracy"])
    return model
```

```python
def get_gan(generator, discriminator):
    adam = get_adam()
    in_gan = Input(shape=(50,))
    layer_1 = generator(in_gan)
    out_gan = discriminator(layer_1)
    
    model = Model(inputs=[in_gan], outputs=[out_gan])
    model.compile(loss="binary_crossentropy", optimizer=adam)
    
    return model
```

```python
generator = get_generator()
discriminator = get_discriminator()
gan = get_gan(generator, discriminator)
batch_size = 2000
```

```python
for i in range(100):
    reals = X[np.random.randint(0, len(X), batch_size)]
    noises = np.random.uniform(size=(batch_size, 50))
    
    fakes = generator.predict(noises)
    
    total = np.vstack([reals, fakes])
    y = np.zeros([2*batch_size, 1])
    y[: batch_size] = 1
    
    discriminator.trainable = True
#     print(total.shape , y.shape)
    discriminator.fit(total , y, verbose=0)

    discriminator.trainable = False
    
    gan.fit(noises, np.ones([batch_size, 1]), verbose=0)
    
    if i%10 == 0:
        print("epoches done", i//10)
        
    plt.figure()
    plt.imshow(fakes[0].reshape(28, 28))
    plt.show()
```
