```python
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("mnist_train_small.csv").values
```

```python
X = data[:, 2:].reshape(-1,28,28,1)
y =data[: ,1]
```

```python
generator = image.ImageDataGenerator(horizontal_flip=True,
                        rotation_range=10,
                        shear_range=.2,
                        zoom_range=.5)
```

```python
gen = generator.flow(X, y)
```

```python
images, values = gen.next()
```
